```python
import pandas as pd

# Create two sample DataFrames
data1 = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C':[1,2,3]
}
df1 = pd.DataFrame(data1)

data2 = {
    'A': [7, 8, 9],
    'B': [10, 11, 12]
}
df2 = pd.DataFrame(data2)

# Concatenate the two DataFrames row-wise
result_row = pd.concat([df1, df2], axis=0)
# Concatenate the two DataFrames column-wise
result_col=pd.concat([df1, df2], axis=1)

print("Concatenated row-wise:")
print(result_row)
print("Concatenated column-wise:")
print(result_col)
```

```python
import pandas as pd
# pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, suffixes=('_x', '_y'), indicator=False, validate=None)
# left: The left DataFrame to be merged.
# right: The right DataFrame to be merged.
# how: The type of merge to be performed. It can be 'inner', 'outer', 'left', or 'right'. The default is 'inner'.
# on: The column or list of columns to merge on. If None, it will use common columns.
# left_on, right_on: Columns to merge on in the left and right DataFrames, respectively.
# left_index, right_index: If True, use the DataFrame's index as the merge key.
# Create two sample DataFrames
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [25, 30, 35, 40]
})

# Merge the two DataFrames on the 'ID' column using inner join
result_inner = pd.merge(df1, df2, on='ID', how='inner')

print("Inner Merge:")
print(result_inner)

```

```python

# In pandas, the join method is used to combine two DataFrames based on their indexes.
import pandas as pd

# Create two sample DataFrames with indexes
left_df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['a', 'b', 'c'])
right_df = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]}, index=['a', 'b', 'd'])
#if have same column in both then give error if any one same then also errrorr

# Perform a left join using join method
join_result = left_df.join(right_df, how='left', lsuffix='_left', rsuffix='_right')

print("Join Result:")
print(join_result)
```
