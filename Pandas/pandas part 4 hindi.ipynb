```python
pd.merge(data1,data2)#default how='inner' work like venn diagram result only the common data bw two dataset pick column that is common in both and according to that give the data of other columns like here it picked key1
```

```python
pd.merge(data1,data2,how = 'left')#pick all left data and pick only common right data 
```

```python
pd.merge(data1,data2,how = 'right')#opposite of the inner=left
```

```python
pd.merge(data1,data2,how = 'outer')#bring data of both columns and see common in them
```
