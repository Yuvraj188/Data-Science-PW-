```python
# trns1=ColumnTransformer([('age_impute',SimpleImputer(strategy='mean'),[3]),
#                         ('embark_impute',SimpleImputer(strategy='most_frequent'),[5])],
#                         remainder='passthrough')

trns1 = ColumnTransformer([
    ('age_impute', SimpleImputer(), [2]),
    ('embark_impute', SimpleImputer(strategy='most_frequent'), [6])
], remainder='passthrough')


```

```python
# trns2=ColumnTransformer([
#     ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,drop='first',handle_unknown='ignore'),[1,6])
# ],remainder='passthrough')
trns2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
# trns2.fit(df)

# Get the number of output columns
# num_columns = trns2.transform(df).shape[1]
# num_columns = trns2.transform(df)
# num_columns
# print("Number of columns after transformation:", num_columns)#want to know no. of columns to apply scaling
```

```python
trns3=ColumnTransformer([('Scaling',StandardScaler(),slice(0,None))])# Scaling all columns
```

```python
trns4=DecisionTreeClassifier()
```

Creating pipeline

```python
pipe=Pipeline([
   ( 'trns1',trns1),
     ('trns2',trns2)
#     ('trns3',trns3),
#     ('trns4',trns4)
])
```

```python
pipe.fit(X_train,y_train)

# If you're doing supervised learning, you would use 'y_train' during model training
# pipe.fit(X_train, y_train)

# Print the transformed data
```
