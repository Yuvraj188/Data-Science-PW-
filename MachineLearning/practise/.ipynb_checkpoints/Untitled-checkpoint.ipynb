```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
df
```

#we will do eda later in this part but first make a pipeline so to revise the steps:
1)impute the missing values 
2)feature construction
3)handle categorical features
4)feature construction
5)feature scaling
6)model training



```python
trns1=ColumnTransformer([('age_impute',SimpleImputer(strategy='mean'),[3]),
                        ('embark_impute',SimpleImputer(strategy='most_frequent'),[5])],
                        remainder='passthrough')




```

```python
trns2=ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[2,5])
],remainder='passthrough')
trns2.fit(df)

# Get the number of output columns
# num_columns = trns2.transform(df).shape[1]
num_columns = trns2.transform(df)
num_columns.to_frame()
# print("Number of columns after transformation:", num_columns)#want to know no. of columns to apply scaling
```

```python
trns3=ColumnTransformer([('Scaling',StandardScaler(),slice(0,None))],remainder='passthrough')# Scaling all columns
```

```python
trns4=DecisionTreeClassifier()
```

Creating pipeline

```python
pipe=Pipeline([
   ( 'trns1',trns1),
    ('trns2',trns2),
    ('trns3',trns3),
    ('trns4',trns4)
])
```

```python
pipe.fit(X_train,y_train)
```
