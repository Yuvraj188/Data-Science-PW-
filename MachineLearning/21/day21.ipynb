## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)#mens which have less age survived and womens with greater a g
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)#small age chidren survived and older people not survived 
```

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Did Not Survive')
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])#in every class how many died and alive 
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
k=titanic.groupby('Embarked')
k.mean()['Survived']*100
# (titanic.groupby('Embarked').mean()['Survived']*100)

# This code calculates the percentage of passengers who survived, grouped by the 'Embarked' column in the Titanic dataset. 
# It shows the survival rate for each port of embarkation.
```

## 6. ClusterMap (Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
